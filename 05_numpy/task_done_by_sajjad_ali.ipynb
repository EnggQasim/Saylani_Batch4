```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

image1 = Image.open('pic1.png')
image_array1= np.array(image1)

image2 = Image.open('pic2.png')
image_array2= np.array(image2)


desired_shape = (300,400)

resized_array1 = image1.resize(desired_shape)
resized_array2 = image2.resize(desired_shape)

combined_image = np.concatenate((resized_array1, resized_array2), axis=1)

plt.imshow(combined_image)
```
