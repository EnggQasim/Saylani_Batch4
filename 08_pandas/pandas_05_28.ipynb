```python
def name_generator(max_len):
    r = np.random.randint(3,max_len,1)[0]
    fc = chr(np.random.randint(65,90,1)[0])
    remaing_letters = np.random.randint(97,122,r)
    remaing_letters = list(map(chr, remaing_letters))
    return fc+"".join(remaing_letters)

print(name_generator(5))

list(map(name_generator,np.random.randint(4,20,3)))
```

```python
def dob_generator(start, end):
    year = np.random.randint(start,end,1)[0]
    month = np.random.randint(1,12,1)[0]
    day = np.random.randint(1,28,1)[0]


    return pd.to_datetime(f'{year}/{month}/{day}')

dob_generator(1970,2017)
```

```python
def grading_system(sub1,sub2,sub3,sub4,sub5):
  total = 500
  obt = sub1 + sub2 + sub3 + sub4 + sub5
  per = obt/total*100
  grade = None

  if per >= 80:
      grade = "A+"
  elif per >= 70:
      grade = "A"
  elif per >= 60:
      grade = "B"
  elif per >= 50:
      grade = "C"
  elif per >= 40:
      grade = "D"
  elif per >= 33:
      grade = "E"
  else:
      grade = "Fail"

  return total, obt, per, grade


grading_system(88,98,88,44,55)
```

```python
state_bank = pd.read_excel("./data.xlsx", 
                          sheet_name="State_bank")
state_bank
```

```python
ubl = pd.read_excel("./data.xlsx", 
                          sheet_name="UBL")
ubl
```

# Inner join

```python
display("Left Table",state_bank)
display("Right Table",ubl)
#     left_df   right_df
df1 = state_bank.merge(ubl,
               left_on="CNIC",
               right_on='nic')

display("Result",df1)
```

# Left Join

```python
display("Left Table",state_bank)
display("Right Table",ubl)
#     left_df   right_df
df1 = state_bank.merge(ubl,
               left_on="CNIC",
               right_on='nic',
                      how='left')

display("Result",df1)
```

# Right Join

```python
display("Left Table",state_bank)
display("Right Table",ubl)
#     left_df   right_df
df1 = state_bank.merge(ubl,
               left_on="CNIC",
               right_on='nic',
                      how='right')# only change how value 

display("Result",df1)
```

# Outter join

```python
display("Left Table",state_bank)
display("Right Table",ubl)
#     left_df   right_df
df1 = state_bank.merge(ubl,
               left_on="CNIC",
               right_on='nic',
                      how='outer')# only change how value 

display("Result",df1)
```

# Cross join

```python
display("Left Table",state_bank)
display("Right Table",ubl)
#     left_df   right_df
df1 = state_bank.merge(ubl,
                      how='cross')# only change how value 

display("Result",df1)
```
